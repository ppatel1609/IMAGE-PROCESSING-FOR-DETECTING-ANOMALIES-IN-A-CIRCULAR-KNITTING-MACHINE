CONTRASTING --> REMOVING PIXELS WITH A CERTAIN RANGE (100-130 IN THIS CASE GIVEN THE HISTOGRAM PLOT) 

```python
import cv2 as cv

def adjust_contrast(image_path, window_name, alpha=1.2, beta=50):
    # Load image in grayscale
    image = cv.imread(image_path, cv.IMREAD_GRAYSCALE)
    
    # Adjust contrast
    adjusted_image = cv.convertScaleAbs(image, alpha=alpha, beta=beta)
    mask = cv.inRange(adjusted_image, 100, 110)
    adjusted_image[mask > 0] = 0
    hist = cv.calcHist([adjusted_image], [0], None, [256], [0,256])
    
    # Plot histogram
    plt.figure(figsize=(8, 4))
    plt.bar(np.arange(256), hist[:,0], width=1)
    plt.title('Intensity Distribution')
    plt.xlabel('Intensity')
    plt.ylabel('Frequency')
    plt.xlim([0, 255])
    plt.grid(True)
    plt.show()
    
    # Display the processed image
    cv.imshow(window_name, adjusted_image)

# Image paths (use your original paths here)
image_path1 = 'Good_image/anomaly1_412.jpg'
image_path2 = 'Minor_default/6.jpg'
cv.imshow('Original image',cv.imread(image_path2))
# Adjust contrast for each image
adjust_contrast(image_path1, 'Good Image')
adjust_contrast(image_path2, 'Bad Image')

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os

def plot_histogram(image, title):
    hist = cv.calcHist([image], [0], None, [256], [0, 256])
    plt.figure(figsize=(8, 4))
    plt.bar(np.arange(256), hist[:, 0], width=1)
    plt.title(title)
    plt.xlabel('Intensity')
    plt.ylabel('Frequency')
    plt.xlim([0, 255])
    plt.grid(True)
    plt.show()

# Load image in grayscale
image_path = os.path.join('Minor_default', '1.jpg')
image = cv.imread(image_path, cv.IMREAD_GRAYSCALE)
image=image[:480,:]
# Original image
cv.imshow('Original Image', image)
plot_histogram(image, 'Original Image')

# Method 1: Histogram Equalization
hist_eq = cv.equalizeHist(image)
cv.imshow('Histogram Equalized', hist_eq)
plot_histogram(hist_eq, 'Histogram Equalized')

# Method 2: CLAHE
clahe = cv.createCLAHE(clipLimit=1.7, tileGridSize=(8, 8))
clahe_img = clahe.apply(image)
cv.imshow('CLAHE', clahe_img)
plot_histogram(clahe_img, 'CLAHE')

# Method 3: Adjusting Brightness and Contrast
alpha = 1.5  # Simple contrast control (1.0-3.0)
beta = 50    # Simple brightness control (0-100)
bright_contrast = cv.convertScaleAbs(image, alpha=alpha, beta=beta)
cv.imshow('Brightness and Contrast Adjusted', bright_contrast)
plot_histogram(bright_contrast, 'Brightness and Contrast Adjusted')

# Method 4: Gamma Correction
gamma = 1.5  # Gamma > 1 will brighten the image, Gamma < 1 will darken the image
look_up_table = np.array([((i / 255.0) ** (1.0 / gamma)) * 255 for i in np.arange(0, 256)]).astype('uint8')
gamma_corrected = cv.LUT(image, look_up_table)
cv.imshow('Gamma Corrected', gamma_corrected)
plot_histogram(gamma_corrected, 'Gamma Corrected')

cv.waitKey(0)
cv.destroyAllWindows()
```
